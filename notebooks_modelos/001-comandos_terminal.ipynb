## Shell de Unix

La terminal, consola, línea de comandos o [Shell](https://es.wikipedia.org/wiki/Shell_de_Unix) de GNU/Linux es, son todos sinónimos de una herramienta poderosa que encontraremos en Linux. La terminal consiste de un intérprete de comandos, que permite realizar tareas de manera más eficiente y rápida que desde un entorno gráfico.

## Comandos basicos de Linux

Abrir la terminal `Ctrl+Alt+T`, seguido al prompt ingresar cualquiera de los siguientes comandos [Molloy D., 2016]:

| Comandos | Descripcion   |  
| :---: | :--- |
| more /etc/isssue          | Distribución utilizada. |
| whoami                    | Usuario logged actualmente.   | 
| uptime                    | Datos del tiempo de encedido. | 
| clear                     | Limpia la terminal. |
| top                       | Lista todos los procesos en ejecución (Crt+C, para salir). |
| uname -a                  | Versión de linux corriendo actualmente. |
| cal                       | Muestra el calendario. |

## Comandos para el sistema de archivos

| Comandos | Nombre (ingles)| Descripcion  |  Ejemplo |
| :---: | :--- | :--- | :--- |
| ls          | List files | Lista los archivos. | ls -alh |
| pwd         | Print the working directory | Directorio de trabajo. | pwd -p |
| cd          | Change directory | Cambia el directorio de trabajo. | cd /home/pi |
| mkdir       | Make a directory | Crea una carpeta. | mkdir test |
| rm          | Remove | Eliminar un archivo o directorios. | rm bad.txt |
| cp          | Copy | Copia un archivo o directorio. | cp a.txt b.txt |
| mv          | Move | Mueve un archivo o directorio. | mv directorio ruta_de_destino/nombre_directorio |
| mv          | Move | Renombrar archivo o directorio. | mv directorio directorio_renombrado |


Se agrego el nombre en ingles, porque en la mayoria de los casos hay una relación entre el nombre y el comando.

> **Nota:** para conocer el __docstring__ del comando y ver las opciones disponibles ademas de la descripción, se agrega `--help` despues del comando, por ejemplo:`ls --help`

## Shortcuts

| Shortcut | Descripción |
| :---: | :--- |
| Flecha arriba      | Devuelve el ultimo comando ingresado. | 
| Tecla Tab          | Autocompletar, directorios, archivos e incluso comandos. | 
| Ctrl+L             | Limpia la terminal. | 
| Ctrl+C             | Para los procesos en ejecución. | 

``` shell
cd /tmp
pwd
mkdir test
cd test
touch hello.txt
ls -l hello.txt
cd ..
cp -r test /tmp/test2
cd /tmp/test2
ls -l
```

> **Nota:** al usar Raspbian / Debian y Ubuntu en las cuentas de usuario, a menudo debe prefijar ciertos comandos con la palabra "sudo". Porque "sudo" es un programa que permite a los usuarios ejecutar programas con privilegios de seguridad del superusuario.

Para conocer [más comandos](https://blog.desdelinux.net/mas-de-400-comandos-para-gnulinux-que-deberias-conocer/) ingrese al link.

## Actualizar el sistema operativo

La distribución de Ubuntu se puede actualizar usando unos pocos pasos. Lo primero es llamar `apt update`, descarga las listas de paquetes de las ubicaciones de Internet identificadas en el archivo "/etc/apt/sources.list". Esto no instala una nueva versión del software; más bien, actualiza las listas de paquetes y sus dependencias. Cuando se complete esta actualización, puede descargar e instalar automáticamente las últimas versiones del software disponible utilizando el comando `apt upgrade`. 

> **Nota:** estos comandos se deben realizar con permisos de administrador (`sudo`).

### Instalar paquetes, programas o aplicaciones en Ubuntu

A continuación vamos a mostrar la instalación por medio de consola, pero existen otras alternativas, para más información ingresar a la siguiente [web](http://www.ubuntu-guia.com/2009/06/como-instalar-paquetes-y-programas-en.html).

Existen dos comandos para instalar programas en modo texto: con "apt" o con "aptitude".
La instalación de software es una tarea administrativa y por ello hay que utilizar el comando "sudo"

> **Nota:** los paquetes con nombres compuestos traen un guión entre las palabras: Ej: compizconfig-settings-manager

* Instalar paquetes
``` shell
sudo apt-get install nombre_paquete
```
*  Instalar varios paquetes a la vez dejando un espacio en blanco entre ellos ej. 
``` shell
sudo apt-get install gnome-do pidgin amsn
```
* Desinstalar paquetes:
``` shell
sudo apt-get remove nombre_paquete
```
* Desinstalar paquetes (incluyendo archivos de configuración):
``` shell
sudo apt-get remove --purge nombre_paquete
```
*  Obtener una lista de opciones del comando:
``` shell
sudo sudo apt-get help
```
> **Nota**: a partir de Ubuntu 14.04 se puede sustituir "apt-get" por simplemente "apt".
